```python
# (a)
import pandas as pd
import numpy as np
# get the 2011-2012data
file=pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2011-2012/DEMO_G.XPT')
file_11_12=pd.DataFrame(file)
file_11_12=file_11_12[["SEQN", "RIDAGEYR", "RIDRETH3", "DMDEDUC2", "DMDMARTL", 
            "RIDSTATR", "SDMVPSU", "SDMVSTRA", "WTMEC2YR", "WTINT2YR"]]
year=[]
total=len(file_11_12.index)
for i in range(total):
    year.append("2011-2012")
file_11_12['year']= year

# get the 2013-2014data
file=pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2013-2014/DEMO_H.XPT')
file_13_14=pd.DataFrame(file)
file_13_14=file_13_14[["SEQN", "RIDAGEYR", "RIDRETH3", "DMDEDUC2", "DMDMARTL", 
            "RIDSTATR", "SDMVPSU", "SDMVSTRA", "WTMEC2YR", "WTINT2YR"]]
year=[]
total=len(file_13_14.index)
for i in range(total):
    year.append("2013-2014")
file_13_14['year']= year

# get the 2015-2016data
file=pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2015-2016/DEMO_I.XPT')
file_15_16=pd.DataFrame(file)
file_15_16=file_15_16[["SEQN", "RIDAGEYR", "RIDRETH3", "DMDEDUC2", "DMDMARTL", 
            "RIDSTATR", "SDMVPSU", "SDMVSTRA", "WTMEC2YR", "WTINT2YR"]]
year=[]
total=len(file_15_16.index)
for i in range(total):
    year.append("2015-2016")
file_15_16['year']= year

# get the 2017-2018data
file=pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2017-2018/DEMO_J.XPT')
file_17_18=pd.DataFrame(file)
file_17_18=file_17_18[["SEQN", "RIDAGEYR", "RIDRETH3", "DMDEDUC2", "DMDMARTL", 
            "RIDSTATR", "SDMVPSU", "SDMVSTRA", "WTMEC2YR", "WTINT2YR"]]
year=[]
total=len(file_17_18.index)
for i in range(total):
    year.append("2017-2018")
file_17_18['year']= year

# combine the file get change the vairbales name 
data_DEMO=pd.concat([file_11_12, file_13_14, file_15_16,file_17_18])
data_DEMO=data_DEMO.rename(columns={"SEQN": "ids", "RIDAGEYR": "age", "RIDRETH3":"race",
                    "DMDEDUC2":"education", "DMDMARTL":"marital"})
data_DEMO=data_DEMO.rename(columns=str.lower)

# change the data type
data_DEMO['ids'] = data_DEMO['ids'].astype('int64')
data_DEMO['age'] = data_DEMO['age'].astype('int64')
data_DEMO['race'] = data_DEMO['race'].astype('category')
data_DEMO['education'] = data_DEMO['education'].astype('category')
data_DEMO['marital'] = data_DEMO['marital'].astype('category')
data_DEMO[['ridstatr','sdmvpsu','sdmvstra','wtmec2yr',
      'wtint2yr']]=data_DEMO[['ridstatr','sdmvpsu','sdmvstra',
                         'wtmec2yr','wtint2yr']].astype('int64')
data_DEMO['year'] = data_DEMO['year'].astype('category')

# save the dataframe as pickle
data_DEMO.to_pickle('DEMO.pkl')
```

```python
# (b)
import pandas as pd
import numpy as np
    
# get the 2011-2012data
file=pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2011-2012/OHXDEN_G.XPT')
file_11_12=pd.DataFrame(file)
cols=list(range(64))
file_11_12=file_11_12[file_11_12.columns[cols]]
file_11_12=file_11_12.drop(columns=['OHDEXSTS', 'OHXIMP'])
year=[]
total=len(file_11_12.index)
for i in range(total):
    year.append("2011-2012")
file_11_12['year']= year

# get the 2013-2014data
file=pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2013-2014/OHXDEN_H.XPT')
file_13_14=pd.DataFrame(file)
cols=list(range(64))
file_13_14=file_13_14[file_13_14.columns[cols]]
file_13_14=file_13_14.drop(columns=['OHDEXSTS', 'OHXIMP'])
year=[]
total=len(file_13_14.index)
for i in range(total):
    year.append("2013-2014")
file_13_14['year']= year

# get the 2015-2016data
file=pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2015-2016/OHXDEN_I.XPT')
file_15_16=pd.DataFrame(file)
cols=list(range(64))
file_15_16=file_15_16[file_15_16.columns[cols]]
file_15_16=file_15_16.drop(columns=['OHDEXSTS', 'OHXIMP'])
year=[]
total=len(file_15_16.index)
for i in range(total):
    year.append("2015-2016")
file_15_16['year']= year

# get the 2017-2018data
file=pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2017-2018/OHXDEN_J.XPT')
file_17_18=pd.DataFrame(file)
cols=list(range(64))
file_17_18=file_17_18[file_17_18.columns[cols]]
file_17_18=file_17_18.drop(columns=['OHDEXSTS', 'OHXIMP'])
year=[]
total=len(file_17_18.index)
for i in range(total):
    year.append("2017-2018")
file_17_18['year']= year

# combine the file get change the vairbales name 
data_OHXDEN=pd.concat([file_11_12, file_13_14, file_15_16,file_17_18])
data_OHXDEN=data_OHXDEN.rename(columns={"SEQN": "ids"})
data_OHXDEN=data_OHXDEN.rename(columns=str.lower)

# change the data type
for column in data_OHXDEN.columns:
    data_OHXDEN[column]=data_OHXDEN[column].astype('category')
for column in data_OHXDEN.columns:
    data_OHXDEN[column]=data_OHXDEN[column].astype('int64', errors='ignore')

# save the dataframe as pickle
data_OHXDEN.to_pickle('OHXDEN.pkl')
```
